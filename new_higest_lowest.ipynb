```python
from collections import defaultdict
from collections.abc import Callable
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
```

```python

dict_day = defaultdict(lambda: defaultdict(lambda: 0))

with open('all.txt') as fp:
    for line in fp.readlines():
        symbol, date, time, price, move = line.split(',')
        move = move.rstrip()
        # print(date, symbol, move, time[:5])
        # dict_day[date][move][time[:5]].append({'symbol': symbol, 'price': price})
        if move == 'higher':
            dict_day[date][time[:5]] += 1
        if move == 'lower':
            dict_day[date][time[:5]] -= 1
```

```python
for x in dict_day:
    print(x)
```

```python
def plot_day(day):
    x = []
    y = []
    for t in sorted(dict_day[day].keys()):
        # print(t, dict_day['2022-07-01'][t])
        x.append(t)
        y.append(dict_day[day][t])

    plt.plot(x, y)
```

```python
plot_day('2022-07-01')
plot_day('2022-06-30')
plot_day('2022-06-21')
```

```python
plot_day('2022-05-27')
plot_day('2022-05-30')
plot_day('2022-05-31')
```
